```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wine = pd.read_csv("1_Visualization\\winequality-all.csv", comment="#")
```

```python
# Investigating the distribution of alcohol variable
sns.histplot(data = wine, x = "alcohol", binwidth = 0.5, color = "skyblue")
plt.title("Distribution of Alcohol Variable")
plt.show()
```

```python
# Comparing the distribution of alcohol variable between two types of wine i.e. red and white
sns.histplot(data = wine, x = "alcohol", hue = "color", element = "step", stat = "density", common_norm = False, alpha = 0.7)
plt.title("Distribution of Alcohol Variable by Wine Type")
plt.show()
```

```python
# Comparing the distribution of alcohol variable in each of possible quality group defined by response variable
sns.histplot(data = wine, x = "alcohol", hue = "response", multiple = "dodge", element = "step", stat = "density", common_norm = False, alpha = 0.7)
plt.title("Distribution of Alcohol Variable by Quality Group")
plt.show()
```

```python
# Percentage of red and white wines within each quality group
quality_counts = wine.groupby(["response", "color"]).size().reset_index(name = "count")
quality_counts["percent"] = quality_counts["count"] / quality_counts["count"].sum() * 100
sns.barplot(data = quality_counts, x = "response", y = "percent", hue = "color")
plt.title("Percentage of Red and White Wines within each Quality Group")
plt.show()
```

```python
sns.scatterplot(data=wine[wine["color"]=="red"], x="fixed.acidity", y="volatile.acidity", alpha=0.5)
plt.xlabel("Fixed Acidity")
plt.ylabel("Volatile Acidity")
plt.title("Relationship Between Fixed and Volatile Acidity")
plt.show()

```

```python
sns.scatterplot(data=wine[wine["color"]=="white"], x="fixed.acidity", y="volatile.acidity", alpha=0.5)
plt.xlabel("Fixed Acidity")
plt.ylabel("Volatile Acidity")
plt.title("Relationship Between Fixed and Volatile Acidity")
plt.show()
```
